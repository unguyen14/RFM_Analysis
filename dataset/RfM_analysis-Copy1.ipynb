```python
df.drop(['increment_id','Year','Month','sales_commission_code','BI Status'], axis=1, inplace=True)
df.columns
```

```python
df.rename(columns={'category_name_1': 'category', 'Working Date': 'working_date', 
                   ' MV ': 'market_value', 'FY' : 'fiscal_year','M-Y': 'M_Y','Customer ID' : 'customer_id'}, inplace=True)
df.columns
```

```python
from datetime import datetime 
#convert working date to datetime
df.working_date= df.working_date.apply( lambda x: datetime.strptime(x, '%m/%d/%Y'))
#convert created at to datetime
df.created_at= df.created_at.apply( lambda x: datetime.strptime(x, '%m/%d/%Y'))
#compare duration to check for duplicated columns
duration = df.working_date - df.created_at
duration.unique()
```

```python
#convert M-Y to datetime
df['M_Y'] = df['M_Y'].apply( lambda x: datetime.strptime(x, '%m-%Y'))
```

```python
df.drop('created_at', axis=1, inplace=True)
df.columns
```

```python
df.isna().sum()
```

```python
skus=df[df.category.isnull()].sku.unique()
for i in skus:
    print(df[df.sku==i])
```

### Comments:
All the Nan values in category do not associate with any specified categories, therefore will be filled with 'Not Identified".

```python
df.category.fillna("Not Identified", inplace = True)
df.loc[df.category == "\\N", "category"] = "Others"
```

```python
df.category.unique()
```

```python
cat_col = df.select_dtypes('object').columns

mask = np.column_stack([df[col].str.contains(r"\\N", na=False) for col in cat_col])
df.loc[mask.any(axis=1)]
```

```python
df.loc[df.payment_method == "cashatdoorstep", "payment_method"] = "cod"
df.loc[df.payment_method == "ublcreditcard", "payment_method"] = "creditcard"
df.loc[df.payment_method == "apg", "payment_method"] = "creditcard"
df.loc[df.payment_method == "mcblite", "payment_method"] = "internetbanking"
df.loc[df.payment_method == "bankalfalah", "payment_method"] = "internetbanking"
df.loc[df.payment_method == "Easypay", "payment_method"] = "e-wallet"
df.loc[df.payment_method == "mygateway", "payment_method"] = "e-wallet"
df.loc[df.payment_method == "jazzwallet", "payment_method"] = "e-wallet"
df.loc[df.payment_method == "Payaxis", "payment_method"] = "e-wallet"
df.loc[df.payment_method == "customercredit", "payment_method"] = "credit/vocher"
df.loc[df.payment_method == "productcredit", "payment_method"] = "credit/vocher"
df.loc[df.payment_method == "easypay_voucher", "payment_method"] = "credit/vocher"
df.loc[df.payment_method == "Easypay_MA", "payment_method"] = "credit/vocher"
df.loc[df.payment_method == "jazzvoucher", "payment_method"] = "credit/vocher"
df.loc[df.payment_method == "marketingexpense", "payment_method"] = "others"
df.loc[df.payment_method == "financesettlement", "payment_method"] = "others"
```
